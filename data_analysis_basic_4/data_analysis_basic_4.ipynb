## 4️⃣ Visualizing data using Matplotlib

### Line plot

```python
#### Basics of line plot
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots() #if subplot has no argument, it draws only one graph.
x = np.arange(15)
y = x ** 2
ax.plot(
    x, y,
    linestyle=":",
    marker="*",
    color="#524FA1"
)
```

```python
#### Line Style
```

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, linestyle="-") #solid
ax.plot(x, x+2, linestyle="--") #dashed
ax.plot(x, x+4, linestyle="-.") #dashdot
ax.plot(x, x+6, linestyle=":") #dotted
```

```python
#### Color
```

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, color="r")
ax.plot(x, x+2, color="green")
ax.plot(x, x+4, color="0.7") #Gray. 1 means white and 0 means black.
ax.plot(x, x+6, color="#524FA1")
```

```python
#### Marker
```

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, marker=".")
ax.plot(x, x+2, marker="o")
ax.plot(x, x+4, marker="v")
ax.plot(x, x+6, marker="s")
ax.plot(x, x+8, marker="*")
```

```python
#### Adjusting xlim & ylim
```

```python
x = np.linspace(0, 10, 1000) #(start, end, step)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_xlim(-2, 12) #sets limit of x axis to -2 and 12.
ax.set_ylim(-1.5, 1.5) #sets limit of y axis to -1.5 and 1.5
```

```python
#### Legend
```

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label='y=x^2')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(loc='upper right',
         shadow=True,
         fancybox=True,
         borderpad=2)
```

```python
### Bar plot
```

```python
#### Basics of Bar plot
```

```python
x = np.arange(10)
fig, ax = plt.subplots(figsize=(12, 4)) #Adjust the aspect ratio of the graph to 12:4.
ax.bar(x, x*2)
```

```python
x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)
data = [x, y, z]
fig, ax = plt.subplots()
x_ax = np.arange(3)
for i in x_ax :
    ax.bar(x_ax, data[i],
          bottom=np.sum(data[:i], axis=0)) #bottom -> bottom of each data
ax.set_xticks(x_ax)
ax.set_xticklabels(["A", "B", "C"])
```

```python
#### Histogram
```

```python
fig, ax = plt.subplots()
data = np.random.randn(1000)
ax.hist(data, bins=50) #bins -> number of bars in histogram
```

```python
### Matplotlib with Pandas
```

```python
df = pd.read_csv("./president_heights.csv")
fig, ax = plt.subplots()
ax.plot(df["order"], df["height(cm)"], label="height")
ax.set_xlabel("order")
ax.set_ylabel("height(cm)")
```

```python
df = pd.read_csv("./pokemon.csv")
fire = df[(df['Type 1'] == 'Fire') | ((df['Type 2']) == "Fire")]
water = df[(df['Type 1'] == 'Water') | ((df['Type 2']) == "Water")]
fig, ax = plt.subplots()
ax.scatter(fire['Attack'], fire['Defense'], color='red', label='Fire', marker="*", s=50)
ax.scatter(water['Attack'], water['Defense'], color='blue', label='Water', s=25)
ax.set_xlabel("Attack")
ax.set_ylabel("Defense")
ax.legend(loc="upper right")
```
